# Extract SIFT files into numpy array for training KMeans

```python
# download
output = "dataset_sift.zip"
if output not in os.listdir():
    url = "https://drive.google.com/uc?id=1_Nj2xlMfAuVl6qy9ODMYEJUIdHru7WDo"
    gdown.download(url, output=output, quiet=False)
    shutil.unpack_archive(filename=output, format="zip")
```

```python
def convert(input_path, output_path, all_desc_h5py):
    """ Read sift files, extract descriptors
    params:
        input_path (str): folders contains 5063 sift files extracted from oxford dataset
        output_path (str): folders contains all descriptors extracted from `filepath`
        all_desc_h5py (str): file hold all descriptors extracted (h5py file)
    """
    input_files = os.listdir(input_path)
    
    if all_desc_h5py in os.listdir():
        os.remove(all_desc_h5py)
    os.makedirs(output_path, exist_ok=True)

    for i, input_file in tqdm(enumerate(input_files)):
        with open(os.path.join(input_path, input_file), 'r') as f:
            text = f.readlines()

            D = int(text[0][:-1]) # dimension descriptors
            N = int(text[1][:-1]) # number of descriptors on image

            if N == 0:
                continue

            values = [value[:-1].split(' ') for value in text[2:]]
            values = np.array(values, dtype=np.float32)

            keys = values[:, :5]
            desc = values[:, 5:].astype(np.uint8)

            with h5py.File(os.path.join(output_path, input_file + ".h5py"), 'w') as hf:
                hf.create_dataset("keypoints", data=keys, dtype="float32")
                hf.create_dataset("descriptors", data=desc, dtype="uint8")

            with h5py.File(all_desc_h5py, 'a') as hf:
                if i == 0:
                    hf.create_dataset("descriptors", data=desc,
                                      chunks=True, maxshape=(None, D))
                else:
                    hf["descriptors"].resize((hf["descriptors"].shape[0] + N), axis=0)
                    hf["descriptors"][-N:] = desc

    with h5py.File(all_desc_h5py, 'r') as hf:
        print("Descriptors shape: {}".format(hf["descriptors"].shape))
```

```python
input_path = "dataset_sift"
output_path = "dataset_sift_h5py"
all_desc_h5py = "all_desc_h5py"

convert(input_path, output_path, all_desc_h5py)

archive_name = "dataset_h5py"
shutil.make_archive(archive_name, format="zip", root_dir=output_path)
shutil.copy(archive_name + ".zip", "/content/drive/My Drive/")

archive_name = "all_desc_h5py"
shutil.make_archive(archive_name, format="zip", root_dir=".", base_dir=all_desc_h5py)
shutil.copy(archive_name + ".zip", "/content/drive/My Drive/")

# file url: 
# "dataset_h5py.zip" https://drive.google.com/uc?id=1509Gvk40omyB4S7u6yiev-pqdBw8137Z
# "all_desc_h5py" https://drive.google.com/uc?id=11rlGbdiuCDkD0o3Q6notumiQgkT6iUK5
```

# Training KMeans on SIFT

```python
# Download descriptors h5py
output = "all_desc_h5py.zip"
if output not in os.listdir():
    url = "https://drive.google.com/uc?id=11rlGbdiuCDkD0o3Q6notumiQgkT6iUK5"
    gdown.download(url, output=output, quiet=False)
    shutil.unpack_archive(filename=output, format="zip")
```

```python
from sklearn.cluster import MiniBatchKMeans

def train(descfile="all_desc_h5py",
          centers_h5py="centers.h5py",
          centers_raw="centers.raw",
          n_iters=50,
          BATCH_SIZE=-1,
          n_clusters=256,
          max_iter=1000):
    """Train KMeans on provided descriptors
    params:
        descfile (str): descriptors file unzipped
        centers_h5py (str): centers saved as h5py format
        centers_raw (str): centers saved by normal way
        n_iters (int): number iterations on training kmeans
        BATCH_SIZE (int): batch_size, default=-1 means use 1/2 training sample
        n_clusters (int): number centers
        max_iter (int): number loops on each batch
    """
    with h5py.File(descfile, 'r') as hf:
        N = hf["descriptors"][:].shape[0]

    if BATCH_SIZE==-1:
        BATCH_SIZE = (N+1)//2

    kmeans = MiniBatchKMeans(
        n_clusters=n_clusters,
        max_iter=max_iter,
        batch_size=BATCH_SIZE//50,
        )

    for i in range(n_iters):
        print("iters: {}".format(i))
        # shuffle
        shuffled_indices = np.random.permutation(N)
        # training on batch
        for start in range(0, N, BATCH_SIZE):
            print('\tbatch: {}'.format(start + 1), end='')
            start_time = time.time()
            end = min(N, start + BATCH_SIZE)
            indices = shuffled_indices[start: end]
            with h5py.File(descfile, 'r') as hf:
                kmeans.partial_fit(hf["descriptors"][:][indices])
            end_time = time.time()
            print('\ttime: {:.2f}'.format(end_time - start_time))
    
    centers = kmeans.cluster_centers_
    # save centers (for python)
    with h5py.File(centers_h5py, 'w') as hf:
        hf.create_dataset("centers", data=kmeans.cluster_centers_, dtype=np.float32)
    
    # save centers (for c++)
    with open(centers_raw, 'w') as f:
        for center in centers:
            center = [str(c) for c in center]
            text = ' '.join(center)
            f.write(text + "\n")
```

```python
descfile = "all_desc.h5py"
centers_h5py = "centers.h5py"
centers_raw = "centers.raw"

# test
#train(n_iters=1, BATCH_SIZE=300)

# train
train(n_iters=50, BATCH_SIZE=-1)

# shutil.copy(centers_h5py, "/content/drive/My Drive/")
# shutil.copy(centers_raw, "/content/drive/My Drive/")
# # file urls:
```

# VLAD

```python
# download centers
output = "centers.h5py"
if output not in os.listdir():
    url = "https://drive.google.com/uc?id=1iUhpbXQxxynxa63D_ernyFrpvnuFu00F"
    gdown.download(url, output, quiet=False)

# download h5py files
output = "dataset_h5py.zip"
if output not in os.listdir():
    url = "https://drive.google.com/uc?id=1509Gvk40omyB4S7u6yiev-pqdBw8137Z"
    gdown.download(url, output=output, quiet=False)
    shutil.unpack_archive(filename=output, extract_dir="dataset_h5py", format="zip")
```

```python
from sklearn.cluster import KMeans
class VLAD:
    def __init__(self, centers, norm=True):
        """VLAD
        args:
            centers (numpy array): Dx128 dimension centers
            norm (bool): use intra normalization
        """
        self.centers = centers
        self.kmeans = KMeans(n_clusters=int(self.centers.shape[0]))
        self.kmeans.cluster_centers_ = self.centers
        self.norm = norm

    def encode(self, desc=None):
        """Encode descriptors into vlad
        params:
            desc (numpy array): Nx8 descriptors of an images
        returns:
            Dx128 numpy array encoded by VLAD
        """
        if desc is None:
            return None
        labels = self.kmeans.predict(desc)
        v = np.zeros(self.centers.shape, dtype=np.float32)
        for i in range(v.shape[0]):
            if np.sum(labels==i) > 0:
                v[i] = np.sum(desc[labels==i,:] - self.centers[i],axis=0)
        if self.norm:
            norms = np.sqrt(np.sum(v**2, axis=1))
            norms[norms < 1e-12] = 1e-12
            v = v / norms.reshape(-1, 1)
        v = v.flatten()
        return v
```

```python
# encode dataset into VLAD (number_data x D x 128)
def encode_dataset(dataset_h5py_path="dataset_h5py",
                   kmeans_centers_path="centers.h5py",
                   vlad_h5py="vlad_h5py",
                   vlad_raw="vlad_raw",
                   vlad_name="vlad_name"):
    """Encode all h5py data into a single vlad file
    params:
        dataet_h5py_path (str): folders contains all h5py files
        kmeans_centers_path (str): centers fil (h5py format)
        vlad_h5py (str): vlad saved as h5py format
        vlad_raw (str): vlad saved as raw
        vlad_name (str): names of all h5py fles
    """
    all_files_h5py = os.listdir(dataset_h5py_path)

    if vlad_h5py in os.listdir():
        os.remove(vlad_h5py)
    if vlad_raw in os.listdir():
        os.remove(vlad_raw)
    if vlad_name in os.listdir():
        os.remove(vlad_name)

    with h5py.File(kmeans_centers_path, 'r') as hf:
        centers = hf["centers"][:]
        vlad = VLAD(centers, norm=True)

    for i, h5py_file in tqdm(enumerate(all_files_h5py)):
        file_name = h5py_file
        #print(file_name)
        full_path = os.path.join(dataset_h5py_path, h5py_file)
        with h5py.File(full_path, 'r') as hf:
            desc = hf["descriptors"][:]
        v = vlad.encode(desc)
        

        with h5py.File(vlad_h5py, 'a') as hf:
            vr = v.reshape(1, -1)
            if i == 0:
                hf.create_dataset("vlads", data=vr, dtype="float32",
                                  chunks=True, maxshape=(None, vr.shape[1]))
            else:
                hf["vlads"].resize((hf["vlads"].shape[0] + 1), axis=0)
                hf["vlads"][-1] = vr

        with open(vlad_raw, 'a') as f:
            strv = [str(vp) for vp in v]
            text = " ".join(strv)
            f.write(text + "\n")
        
        with open(vlad_name, 'a') as f:
            text = file_name.split('.')[0]
            f.write(text + "\n")
```

```python
dataset_h5py_path = "dataset_h5py"
kmeans_centers_path = "centers.h5py"
vlad_h5py = "vlad_h5py"
vlad_raw = "vlad_raw"
vlad_name = "vlad_name"

encode_dataset()


shutil.copy(vlad_h5py, '/content/drive/My Drive/')
shutil.copy(vlad_raw, '/content/drive/My Drive/')
shutil.copy(vlad_name, '/content/drive/My Drive/')

# urls
# vlad_h5py: https://drive.google.com/uc?id=1-5DyU_gWI9MgddKiiw45guZpZLfh0s8T
# vlad_raw: https://drive.google.com/uc?id=1-Dx44sLkqgOi6fctS4uHGpnuz2G3r2pv
# vlad_name: https://drive.google.com/uc?id=1-GqxWpJw6WUmW13wmE1VOtfByMycb-hU
```

# Product Quantization

```python
from sklearn.cluster import KMeans

class PQ:
    def __init__(self, n_clusters=256, dim=16, centers=None):
        """Product quantization for better cluster
        args:
            n_clusters (int): number clusters on each product
            dim (int): dimension of each cluster
        """
        self.n_clusters = n_clusters
        self.dim = dim
        self.centers = centers

    def train(self, V):
        """Train centers with kmeans
        params:
            V (array): dataset's vlad (NxD)
        """
        N, D = V.shape # 5063x32768 (=5063x256x128)
        
        # number of products
        assert D % self.dim == 0
        self.nprods = D // self.dim

        self.centers = np.zeros((self.n_clusters, D), dtype=np.float32)
        for prod in tqdm(range(0, D, self.dim)):
            kmeans = KMeans(n_clusters=self.n_clusters,
                            max_iter=10000,
                            random_state=42)
            kmeans.fit(V[:, prod:prod+self.dim])
            self.centers[:, prod:prod+self.dim] = kmeans.cluster_centers_
    
    def predict(self, V):
        """Predict an array vlad-ed
        params:
            V (array): a MxD array vlad-ed need to be predict
        returns:
            labels (array): a Mx(D/dim) array label
        """
        assert V.shape[1] == self.centers.shape[1]
        kmean = KMeans(n_clusters=self.n_clusters)
        labels = np.zeros((V.shape[0], V.shape[1]//self.dim), dtype=np.uint8)
        for i in range(0, V.shape[1], self.dim):
            kmean.cluster_centers_ = self.centers[:, i:i+self.dim]
            label = kmean.predict(V[:, i:i+self.dim])
            labels[:, i//self.dim] = label
        return labels

    def reconstruct(self, labels):
        """
        Reconstruct an array (predicted) by centers
        This method is only for testing (takes RAM to store values)
        params:
            labels (array): a Mx(D/dim) array contains label need to be reconstructed
        returns:
            V (array): a MxD array (vlad-ed like)
        """
        V = np.zeros((labels.shape[0], labels.shape[1]*self.dim), dtype=np.float32)
        for i in range(0, labels.shape[1]*self.dim, self.dim):
            V[:, i:i+self.dim] = self.centers[:, i: i+self.dim][labels[:, i//self.dim]]
        return V

    def quantize(self, V):
        labels = self.predict(V)
        V = self.reconstruct(labels)
        return V
        
    def save(self, h5py_path, raw_path):
        """Save centers into file
        params:
            h5py_path (str): where to save centers with h5py format
                            `dim`: dimension
                            `centers`: centers
            raw_path (str): save centers with raw format.
                            first line: dim
                            next lines: centers

        """
        with h5py.File(h5py_path, 'w') as hf:
            hf.create_dataset("dim", data=[self.dim], dtype="uint16")
            hf.create_dataset("centers", data=self.centers, dtype="float32")
        with open(raw_path, 'w') as f:
            f.write('{}\n'.format(self.dim))
            for center in self.centers:
                strc = [str(c) for c in center]
                f.write(' '.join(strc) + '\n')
```

```python
# download vlad h5py data
output = "vlad_h5py"
if output not in os.listdir():
    url = "https://drive.google.com/uc?id=1-5DyU_gWI9MgddKiiw45guZpZLfh0s8T"
    gdown.download(url, output, quiet=False)

# creatte product quantization
n_clusters = 256
dim = 64
pq = PQ(n_clusters=n_clusters, dim=dim, centers=None)

with h5py.File(output, 'r') as hf:
    pq.train(hf['vlads'][:])


h5py_path = "pq_centers.h5py"
raw_path = "pq_centers.raw"
pq.save(h5py_path, raw_path)

shutil.copy(h5py_path, '/content/drive/My Drive')
shutil.copy(raw_path, '/content/drive/My Drive')

# urls
# "raw" https://drive.google.com/uc?id=1-Kfk4oV1ZEE3JhpdmRU6ts1V3uSB_HCT
# "h5py" https://drive.google.com/uc?id=1-GrfteYsDy1wWJpib6_lK8EDLiZigplV
```
